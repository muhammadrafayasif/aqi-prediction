## What are the trends in this dataset?

> **Note:** The dataset includes an approximately 5 day gap due to the HopsWorks outage which occurred on October, this is handled in the training pipeline

### 1. Connect to HopsWorks

```python
import hopsworks, os
project = hopsworks.login(api_key_value=os.getenv("API_KEY"))
fs = project.get_feature_store()
fg = fs.get_or_create_feature_group("aqi_feature_pipeline", version=1)
df = fg.read()
```

### 2. Create a line plot of the dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.lineplot(df, x="timestamp", y="aqi")
```
