## What is the earliest and latest AQI recorded?

> **Note:** The data in this notebook is most likely out of date, run it again to get latest data


### 1. Connect to HopsWorks

```python
import hopsworks, os
project = hopsworks.login(api_key_value=os.getenv("API_KEY"))
fs = project.get_feature_store()
fg = fs.get_or_create_feature_group("aqi_feature_pipeline", version=1)
df = fg.read()
```

### 2. Get most recent and latest record

```python
print("Earliest record:\n", df.iloc[0], "\n")
print("Latest record:\n", df.iloc[-1])
```

### 3. Difference in time

```python
df.iloc[-1]["timestamp"] - df.iloc[0]["timestamp"]
```
